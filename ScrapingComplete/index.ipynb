# Code chính

```python
# Import thư viện
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep

import pandas as pd
import json
```

# Đoạn code cào từ đầu

```python

# Đường dẫn tới file JSON
json_file_path = 'vocab_frequency_new.json'
output_json_file_path = 'Vocab_scraping_official.json'

# Định nghĩa các danh mục từ loại
categories_mapping = {
    "Danh từ": "category1: 'Danh từ'",
    "Động từ": "category2: 'Động từ'",
    "Tính từ": "category3: 'Tính từ'",
    "Số từ": "category4: 'Số từ'",
    "Lượng từ": "category5: 'Lượng từ'",
    "Phụ từ": "category6: 'Phụ từ'",
    "Đại từ": "category7: 'Đại từ'"
}
# Đọc dữ liệu từ file JSON
with open(json_file_path, 'r', encoding='utf-8') as file:
    vocab_data = json.load(file)

# Khởi tạo WebDriver với chế độ ẩn
# chrome_options = Options()
# chrome_options.add_argument("--headless")  # Chạy Chrome ở chế độ ẩn (không hiển thị cửa sổ trình duyệt)
# browser = webdriver.Chrome(options=chrome_options)

# Khởi tạo WebDriver (giả sử bạn đang dùng Chrome), nếu như cần chạy hiện rõ trình duyệt thì chạy code ở dưới
browser = webdriver.Chrome()

# URL của trang web
url = "http://tratu.soha.vn/dict/vn_vn/"
browser.get(url)

# Khởi tạo danh sách kết quả
results = []

# Thực hiện tìm kiếm cho từng từ trong file JSON
for key, word in json.load(open(json_file_path, 'r', encoding='utf-8')).items():
    # Lấy lại các phần tử input và button trước mỗi lần tìm kiếm
    input_search = browser.find_element(By.ID, 'search')
    search_button = browser.find_element(By.ID, 'btnSearch')
    
    # Clear the input field
    input_search.clear()
    
    # Send the input to the webpage
    input_search.send_keys(word)
    sleep(1)
    search_button.click()
    
    # Scrap the data
    elements = browser.find_elements(By.XPATH, "//h3/span[@class='mw-headline']")
    
    # Lấy text của tất cả các phần tử và in ra
    word_categories = []
    for element in elements:
        text = element.text.strip()
        if text in categories_mapping:
            word_categories.append(categories_mapping[text])
        elif text not in categories_mapping:
            word_categories.append([])

    # Tạo đối tượng kết quả và thêm vào danh sách
    result = {
        "title": word,
        "categories": word_categories
    }
    results.append(result)
    
    # Lưu kết quả vào file JSON ngay sau khi tìm kiếm mỗi từ
    with open(output_json_file_path, 'w', encoding='utf-8') as outfile:
        json.dump(results, outfile, ensure_ascii=False, indent=4)
    
    print(f"Results for '{word}': {word_categories}")
    print('-' * 20)
    
    # Đợi một chút trước khi thực hiện tìm kiếm từ tiếp theo (tùy chọn)
    sleep(2)

# Đóng trình duyệt
browser.quit()

print(f"Final results have been saved to {output_json_file_path}")
```

# Tiếp tục cào khởi tạo từ index cho cần

```python
# Đường dẫn tới file JSON
json_file_path = 'vocab_frequency_new.json'
output_json_file_path = 'Vocab_scraping_official.json'

# Định nghĩa các danh mục từ loại
categories_mapping = {
    "Danh từ": "category1: 'Danh từ'",
    "Động từ": "category2: 'Động từ'",
    "Tính từ": "category3: 'Tính từ'",
    "Số từ": "category4: 'Số từ'",
    "Lượng từ": "category5: 'Lượng từ'",
    "Phó từ": "category6: 'Phó từ'",
    "Đại từ": "category7: 'Đại từ'",
    "": "category8: 'Null'"
}

# Khởi tạo WebDriver (giả sử bạn đang dùng Chrome), nếu như cần chạy hiện rõ trình duyệt thì chạy code ở dưới
browser = webdriver.Chrome()

# URL của trang web
url = "http://tratu.soha.vn/dict/vn_vn/"
browser.get(url)

# Đọc dữ liệu từ file JSON
with open(json_file_path, 'r', encoding='utf-8') as file:
    vocab_data = json.load(file)

# Tìm vị trí bắt đầu tìm kiếm (ví dụ từ vị trí 33)
start_index = 39

# Khởi tạo danh sách kết quả
results = []

# Thực hiện tìm kiếm từ từ vị trí start_index đến hết danh sách
for index in range(start_index, 1001):
    word = vocab_data[str(index)]
    
    # Lấy lại các phần tử input và button trước mỗi lần tìm kiếm
    input_search = browser.find_element(By.ID, 'search')
    search_button = browser.find_element(By.ID, 'btnSearch')
    
    # Clear the input field
    input_search.clear()
    
    # Send the input to the webpage
    input_search.send_keys(word)
    sleep(1)
    search_button.click()
    
    # Scrap the data
    elements = browser.find_elements(By.XPATH, "//h3/span[@class='mw-headline']")
    
    # Lấy text của tất cả các phần tử và in ra
    word_categories = []
    for element in elements:
        text = element.text.strip()
        if text in categories_mapping:
            word_categories.append(categories_mapping[text])
    
    # Tạo đối tượng kết quả và thêm vào danh sách
    result = {
        "title": word,
        "categories": word_categories
    }
    results.append(result)
    
    # Mở file JSON để thêm mới kết quả
    with open(output_json_file_path, 'a', encoding='utf-8') as outfile:
        json.dump(result, outfile, ensure_ascii=False, indent=4)
        outfile.write(",\n")  # Thêm dấu phẩy và xuống dòng để đảm bảo JSON đúng định dạng
    
    print(f"Results for '{word}': {word_categories}")
    print('-' * 20)
    
    # Đợi một chút trước khi thực hiện tìm kiếm từ tiếp theo (tùy chọn)
    sleep(2)

# Đóng trình duyệt
browser.quit()

print(f"Final results have been saved to {output_json_file_path}")
```

# Đoạn code convert file json về định dạng yêu cầu

```python
import json

# Đường dẫn tới file JSON đầu vào
input_json_file_path = 'vocab_not_null_pre.json'

# Đường dẫn tới file JSON đầu ra
output_json_file_path = 'vocab_not_null.json'

# Hàm chuyển đổi định dạng categories
def convert_categories(categories):
    return [category.split(":")[1].strip().strip("'") for category in categories]

# Đọc dữ liệu từ file JSON đầu vào
with open(input_json_file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)

# Chuyển đổi định dạng categories cho từng đối tượng
for item in data:
    item['categories'] = convert_categories(item['categories'])

# Ghi dữ liệu vào file JSON đầu ra
with open(output_json_file_path, 'w', encoding='utf-8') as file:
    json.dump(data, file, ensure_ascii=False, indent=4)

print(f"Data has been converted and saved to {output_json_file_path}")
```

# Đoạn code loại bỏ data rỗng và bỏ vào vocab_null.json

```python
import json

# Đường dẫn tới file JSON đầu vào
input_json_file_path = 'vocab_scraping_complete_converted.json'

# Đường dẫn tới file JSON data không rỗng
output_json_file_path = 'vocab_not_null.json'

# Đường dẫn tới file JSON đầu ra các mục đó rỗng
null_json_file_path = 'vocab_null.json'

# Hàm chuyển đổi định dạng categories
def convert_categories(categories):
    converted_categories = []
    for category in categories:
        if ':' in category:
            converted_categories.append(category.split(":")[1].strip().strip("'"))
    return converted_categories

# Đọc dữ liệu từ file JSON đầu vào
with open(input_json_file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)

# Khởi tạo danh sách cho các mục có danh sách rỗng và không rỗng
data_with_categories = []
data_without_categories = []

# Chuyển đổi định dạng categories cho từng đối tượng và phân loại các mục
for item in data:
    if item['categories']:
        item['categories'] = convert_categories(item['categories'])
        data_with_categories.append(item)
    else:
        data_without_categories.append(item)

# Ghi dữ liệu vào file JSON đầu ra (các mục không có danh sách rỗng)
with open(output_json_file_path, 'w', encoding='utf-8') as file:
    json.dump(data_with_categories, file, ensure_ascii=False, indent=4)

# Ghi dữ liệu vào file JSON đầu ra (các mục có danh sách rỗng)
with open(null_json_file_path, 'w', encoding='utf-8') as file:
    json.dump(data_without_categories, file, ensure_ascii=False, indent=4)

print(f"Non-empty categories data has been saved to {output_json_file_path}")
print(f"Empty categories data has been saved to {null_json_file_path}")
```

# Đoạn code loại bỏ data rỗng và ghi vào file not_null.json

```python
import json

# Đường dẫn tới file JSON đầu vào
input_json_file_path = 'vocab_scraping_complete_converted.json'

# Đường dẫn tới file JSON đầu ra
output_json_file_path = 'vocab_not_null.json'

# Đọc dữ liệu từ file JSON đầu vào
with open(input_json_file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)

# Lọc bỏ các đối tượng có "categories" là danh sách rỗng
filtered_data = [item for item in data if item['categories']]

# Ghi dữ liệu đã lọc vào file JSON đầu ra
with open(output_json_file_path, 'w', encoding='utf-8') as file:
    json.dump(filtered_data, file, ensure_ascii=False, indent=4)

print(f"Filtered data has been saved to {output_json_file_path}")
```
